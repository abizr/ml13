## prepare model

```python
#label Encoder
from sklearn.preprocessing import LabelEncoder
```

```python
lecdr = LabelEncoder()

df_cancer['diagnosis'] = lecdr.fit_transform(df_cancer['diagnosis'])
```

```python
#memisahkan variabel independen dan variabel dependen
X = df_cancer.drop('diagnosis',axis=1).values
y = df_cancer['diagnosis'].values
```

```python
#split data
from sklearn.model_selection import train_test_split
```

```python
random_state = 44

x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=random_state)
```

## feature scalling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()

X_train = sc.fit_transform(x_train)
X_test= sc.transform(x_test)
```

## modelling

```python
#logistic regression model
from sklearn.linear_model import LogisticRegression
```

```python
logreg= LogisticRegression()

logreg.fit(X_train, y_train)

y_pred_logreg = logreg.predict(X_test)
```

```python
# Random Forest Classifier model
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()

rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)
```

```python
# Decision Tree Classifier model
from sklearn.tree import DecisionTreeClassifier
```

```python
dct = DecisionTreeClassifier()

dct.fit(X_train, y_train)

y_pred_dt = dct.predict(X_test)
```

```python
# Support Vector machine classifier model
from sklearn.svm import SVC
```

```python
svm = SVC(probability=True)

svm.fit(X_train,y_train)

y_pred_svm = svm.predict(X_test)
```

```python
X_train.shape, y_train.shape,X_test.shape, y_test.shape
```

## Evaluation and comparison of all the models

```python
from sklearn.metrics import accuracy_score
```

```python
models = []

Z = [SVC() , DecisionTreeClassifier() , LogisticRegression() , RandomForestClassifier()]


X = ["SVC" , "DecisionTreeClassifier" , "LogisticRegression" , "RandomForestClassifier"]

for i in range(0,len(Z)):
    model = Z[i]
    model.fit( X_train , y_train )
    pred = model.predict(X_test)
    models.append(accuracy_score(pred , y_test))
```

```python
#model accuracy & algorithm
maa = { "Accuracy" : models , "Algorithm" : X }
df = pd.DataFrame(maa)
df
```

```python
sns.barplot(df['Accuracy'],df['Algorithm'],palette= "husl").set_title('total accuracy of ML algorithm model')
```

```python
# Receiver operating characteristic (ROC)
from sklearn.metrics import roc_auc_score,auc,f1_score
from sklearn.metrics import precision_recall_curve,roc_curve
```

```python
y_score = svm.decision_function(X_test)

FPR, TPR, _ = roc_curve(y_test, y_score)
ROC_AUC = auc(FPR, TPR)
print (ROC_AUC)

plt.figure(figsize =[11,9])
plt.plot(FPR, TPR, label= 'ROC curve(area = %0.2f)'%ROC_AUC, linewidth= 4)
plt.plot([0,1],[0,1], 'k--', linewidth = 4)
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('False Positive Rate', fontsize = 18)
plt.ylabel('True Positive Rate', fontsize = 18)
plt.title('Receiver operating characteristic', fontsize= 18)
plt.show()
```
